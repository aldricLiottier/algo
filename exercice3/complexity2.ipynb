```python
from matplotlib import pyplot as plt
import numpy as np
import time

def function_2(n: int) -> None:
    print(n)
    for i in range(n):
        temp_list = [j+i for j in range(n)]
        shuffle(temp_list)
        max(temp_list)

input_range = 150
time_list = []

for i in range (input_range):
    start_time = time.time()
    function_2(i)
    end_time = time.time()
    time_list.append((end_time - start_time))

x = np.array(range(1,input_range+1))
y = time_list
z = np.polyfit(x, y, 3)

p = np.poly1d(z)
p30 = np.poly1d(np.polyfit(x, y, 30))

xp = np.linspace(0, input_range)
_ = plt.plot(x, y, '.', xp, p(xp), '-', xp, p30(xp), '--')

plt.xlabel("n")
plt.ylabel("time to compute f(n) (seconds)")
plt.show()
```
