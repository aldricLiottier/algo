```python
from matplotlib import pyplot as plt
import numpy as np
import time

def function_1(n: int) -> None:
    temp_list = list()
    for i in range(n**2):
        temp = 0
        for j in range(i):
            temp += j
        temp_list.append(temp)
    sum(temp_list)

input_range = 50
time_list = []

for i in range (input_range):
    start_time = time.time()
    function_1(i)
    end_time = time.time()
    time_list.append((end_time - start_time))

x = np.array(range(1,input_range+1))
y = time_list
z = np.polyfit(x, y, 3)

p = np.poly1d(z)
p30 = np.poly1d(np.polyfit(x, y, 30))

xp = np.linspace(0, input_range)
_ = plt.plot(x, y, '.', xp, p(xp), '-', xp, p30(xp), '--')

plt.xlabel("n")
plt.ylabel("time to compute f(n) (seconds)")
plt.show()
```
