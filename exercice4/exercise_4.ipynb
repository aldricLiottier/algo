## 1 - Definitions

### Line graph

Given a graph G with V vertices and E edges, a line graph of G is a graph L(G) with |E| vertices where each vertex represents an edge of G. Two vertices in L(G) are connected if and only if the corresponding edges in G share a vertex.

### Eulerian path

A eulerian path is a path that visits each edge exactly once (vertices can be visited several times.)

### Hamiltonian path

A hamiltonian path is a path that visits each vertex exactly once.

## Exercise

When trying to find an eulerian path in a graph G, the goal is to find a path that visits each edge exactly once. This is equivalent to finding a hamiltonian path in the line graph L(G). This is because each edge in G corresponds to a vertex in L(G) and each vertex in G corresponds to an edge in L(G). Therefore, finding a hamiltonian path in L(G) is equivalent to finding an eulerian path in G.

## 2 - Code examples


```python
import networkx as nx

# Create a graph
N = 5
G = nx.tournament.random_tournament(N)

def visualise_graph(G: nx.DiGraph, **kwargs) -> None:
    """Visualise a graph using networkx and matplotlib"""
    pos = nx.spring_layout(G, k=0.8, iterations=30)  # Makes it less cluttered
    nx.draw(G, pos=pos, with_labels=True, **kwargs)


visualise_graph(G)
```

```python
# Compute the line graph
LG = nx.line_graph(G)

visualise_graph(LG)
```

```python
# Find eulerian path in graph
if nx.has_eulerian_path(G):
    EP = nx.eulerian_path(G)
    print(list(EP))
else:
    print('No eulerian path found')
```

```python
# Find a Hamiltonian path in the line graph
HP = nx.tournament.hamiltonian_path(LG)
print(list(HP)) # These results are not correct

edges = [(HP[i], HP[i + 1]) for i in range(len(HP)-1)]

edge_color = ['red' if edge in edges else 'black' for edge in LG.edges()]
visualise_graph(LG, edge_color=edge_color)
```
